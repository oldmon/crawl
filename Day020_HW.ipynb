```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
d={}
i=0
for tr in table.find_all('tr'):
    if i==0:
        i=1
        continue
    elif i%12==1:
        poll=tr.find_all('td')[0].get_text()
        d.setdefault(poll,{})
        month=tr.find_all('td')[2].get_text()
        value=tr.find_all('td')[3].get_text()
    else:
        month=tr.find_all('td')[0].get_text()
        value=tr.find_all('td')[1].get_text()
    d[poll][month]=value
    i+=1
```
