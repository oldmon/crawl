# Ettoday 網路爬蟲實作練習


* 能夠利用 Request + BeatifulSour 撰寫爬蟲，並存放到合適的資料結構


## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
import requests,time
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import Select

browser = webdriver.Chrome()
browser.get("https://www.ettoday.net/news/news-list.htm")
for i in range(10):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 10000);")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    if datetime.strptime(d.find(class_="date").text,'%Y/%m/%d %H:%M')<datetime.now()-timedelta(hours=2):
        break
    else:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
targetdate=datetime.now()-timedelta(days = 3)
selectyr=Select(browser.find_element_by_id('selY'))
selectyr.select_by_value(str(targetdate.year))
selectmon=Select(browser.find_element_by_id('selM'))
selectmon.select_by_value(str(targetdate.month))
selectday=Select(browser.find_element_by_id('selD'))
selectday.select_by_value(str(targetdate.day))
browser.find_element_by_id('button').click()

for i in range(48):
    time.sleep(1)
    browser.execute_script('window.scrollTo(0, '+str(512*i)+');')
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if datetime.strptime(d.find(class_="date").text,'%Y/%m/%d %H:%M')>datetime(targetdate.year,targetdate.month,targetdate.day,17):
        continue
    elif datetime.strptime(d.find(class_="date").text,'%Y/%m/%d %H:%M')<datetime(targetdate.year,targetdate.month,targetdate.day,15):
        break
    else:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)

browser.quit()
```
