### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
newsInCat={}

for d in soup.find(class_="part_list_2").find_all('h3'):
    typ=d.find('em').text
    newsInCat.setdefault(typ, [])
    newsInCat[typ].append({'time':d.find(class_="date").text,'title':d.find_all('a')[-1].text})

print(newsInCat)
```

### ③ 哪一個類別的文章最多

```python
keylist=[]
num=[]
for key in newsInCat.keys():
    keylist.append(key)
    num.append(len(newsInCat[key]))

print(keylist[num.index(max(num))])
```
